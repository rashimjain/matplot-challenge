```python
#1- Import Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
import seaborn as sns
```

```python
#2- Reading File 1 - Clinical Trial data
clinic_trial_file_path = os.path.join('raw_data','clinicaltrial_data.csv')
trial_data_pd = pd.read_csv(clinic_trial_file_path)
print (trial_data_pd.head())

total_data1 = trial_data_pd.count(axis=0)
print ("--------------------------------------")
print (total_data1)
```

```python
#2- Reading File 2 - Mouse drug data.csv
mouse_drug_file_path = os.path.join('raw_data','mouse_drug_data.csv')
drug_data_pd = pd.read_csv(mouse_drug_file_path)
print (drug_data_pd.head())

total_data2 = drug_data_pd.count(axis=0)
print ("--------------------------------------")
print (total_data2)
```

```python
#4- Combine both data sets
pyma_data = trial_data_pd.merge(drug_data_pd, on = 'Mouse ID', how = 'outer')
pyma_data.head()
```

```python
#5 Getting Number of Drugs Data
df1 = pyma_data.groupby(["Drug"])["Mouse ID"].count()
df1
```

```python
#6 Calculating average tumor volume by drug
average_volume_tumor = pyma_data.groupby(["Drug","Timepoint"])["Tumor Volume (mm3)"].mean()
average_volume_tumor
```

```python
#7 Calculating variance in Tumor Volume by drug - using "pandas.DataFrame.sem() - 
# that returns unbiased standard error of the mean over requested axis
# https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.sem.html

volume_error_tumor = pyma_data.groupby(["Drug","Timepoint"])["Tumor Volume (mm3)"].sem()
volume_error_tumor.head()
```

```python
# Unstacking Average Tumor Volume data.
# http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.unstack.html
# https://pandas.pydata.org/pandas-docs/stable/reshaping.html

average_volume_tumor = average_volume_tumor.unstack(0)
average_volume_tumor
```

```python
# Unstacking Volume error data.
volume_error_tumor = volume_error_tumor.unstack(0)
volume_error_tumor
```

```python
#8 - Plotting the first scatter plot on - how the tumor volume changes over time for each treatment
# https://matplotlib.org/api/_as_gen/matplotlib.pyplot.errorbar.html


# Listof drug names -- the list function provides the column names
# drug = list (average_volume_tumor)
drug = drug_data_pd['Drug'].unique()
drug.sort()
print (drug)

# markers for scatter plot - we will be plotting only 4 drugs
# markers = ['o', '^', 's', 'd']
markers = ['o', '^', 's', 'd','o', '^', 's', 'd','o', '^']

# xaxis is the Timepoint
xaxis = average_volume_tumor.index

# Other plot parameters
plt.figure(figsize = (15,10))
plt.title('Tumor Volume Over Time', fontdict = {'fontsize': 20, 'fontweight': 'bold'})
plt.xlabel('Time (Days)', fontdict = {'fontsize': 14})
plt.ylabel('Tumor Volume (mm3)', fontdict = {'fontsize': 14})
plt.xlim(0, 45) 

# Plotting the scatter plot
for i in range (0,10):
#for i in range (0,4):
    plt.errorbar(xaxis, average_volume_tumor[drug[i]], volume_error_tumor[drug[i]], 
                 linestyle = '--', marker =  markers[i], capthick = 1, capsize = 3)

#legend options
legend = plt.legend(numpoints = 2,frameon = True, markerscale = 1, 
                edgecolor = 'black', fontsize = '14', framealpha = 1)

# Setting theme using Seaborne library function
sns.set_style('ticks')

plt.savefig("Images/Chart-1-Errorbar_Plot_Tumor_Volume.png")

plt.show()
```

```python
#9 Calculating average and error rate of Metastatic Sites by drug
average_volume_meta = pyma_data.groupby(["Drug","Timepoint"])["Metastatic Sites"].mean()
print (average_volume_meta.head())

volume_error_meta = pyma_data.groupby(["Drug","Timepoint"])["Metastatic Sites"].sem()
print (volume_error_meta.head())
```

```python
# Unstacking Metastatic Sites by drug
# http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.unstack.html

average_volume_meta = average_volume_meta.unstack(0)
print (average_volume_meta)

volume_error_meta = volume_error_meta.unstack(0)
print (volume_error_meta)
```

```python
#10- Plotting the second scatter plot on - how the number of metastatic/ cancer spreading sites changes over time for each treatment.
# https://matplotlib.org/api/_as_gen/matplotlib.pyplot.errorbar.html


# Listof drug names -- the list function provides the column names
# drug = list (average_volume_tumor)
drug = drug_data_pd['Drug'].unique()
drug.sort()
print (drug)

# markers for scatter plot - we will be plotting only 4 drugs
# markers = ['o', '^', 's', 'd']
markers = ['o', '^', 's', 'd','o', '^', 's', 'd','o', '^']

# xaxis is the Timepoint
xaxis = average_volume_meta.index

# Other plot parameters
plt.figure(figsize = (15,10))
plt.title('Metastatic Spread During Treatment', fontdict = {'fontsize': 20, 'fontweight': 'bold'})
plt.xlabel('Treatment Duration (Days)', fontdict = {'fontsize': 14})
plt.ylabel('Metastatic Sites)', fontdict = {'fontsize': 14})
plt.xlim(0, 45) 

# Plotting the scatter plot
for i in range (0,10):
#for i in range (0,4):
    plt.errorbar(xaxis, average_volume_meta[drug[i]], volume_error_meta[drug[i]], 
                 linestyle = '--', marker =  markers[i], capthick = 1, capsize = 3)

#legend options
legend = plt.legend(numpoints = 2,frameon = True, markerscale = 1, 
                edgecolor = 'black', fontsize = '14', framealpha = 1)

# Setting theme using Seaborne library function
sns.set_style('ticks')

plt.savefig("Images/Chart-2-Errorbar_Plot_Metastatic_Spread.png")

plt.show()
```

```python
#11 - Calculating Survival rate - number of mice still alive through the course of treatment

survival_data = pyma_data.groupby(["Drug","Timepoint"])["Mouse ID"].count()
survival_data = survival_data.unstack(0)
print (survival_data)
```

```python
y = survival_data.loc[0,"Propriva"]
print (y)
```

```python
#12 - Scatter plot that shows the number of mice still alive through the course of treatment (Survival Rate)

# markers for scatter plot - we will be plotting only 4 drugs
# markers = ['o', '^', 's', 'd']
markers = ['o', '^', 's', 'd','o', '^', 's', 'd','o', '^']

# xaxis is the Timepoint
xaxis = survival_data.index

# Other plot parameters
plt.figure(figsize = (15,10))
plt.title('Survival During Treatment', fontdict = {'fontsize': 20, 'fontweight': 'bold'})
plt.xlabel('Time (Days)', fontdict = {'fontsize': 14})
plt.ylabel('Survival rate (%)', fontdict = {'fontsize': 14})
plt.xlim(0, 45) 

# Plotting the scatter plot
for i in range (0,10):
#for i in range (0,4):
    yaxis = survival_data[drug[i]]/survival_data.loc[0,drug[i]]
    yaxis = yaxis * 100
    plt.plot(xaxis, yaxis, linestyle = '--', marker =  markers[i])

#legend options
legend = plt.legend(numpoints = 2,frameon = True, markerscale = 1, 
                edgecolor = 'black', fontsize = '14', framealpha = 1)

# Setting theme using Seaborne library function
sns.set_style('ticks')

plt.savefig("Images/Chart-3-Line_Chart_Survival_Rate.png")

plt.show()
```

```python
#13 - Creating a bar graph that compares the total % tumor volume change for each drug across the full 45 days.
print (average_volume_tumor)
print (drug)
```

```python
# Estimating Tumor Change
tumor_change =[]

for j in range (0,10):
    avg = (average_volume_tumor.loc[45, drug[j]] /  average_volume_tumor.loc[0, drug[j]]) - 1
    avg = round (avg * 100,2)
    tumor_change.append(avg)
print (tumor_change)

x = len (drug)
xaxis = np.arange(1,x+1)

print (x)

# Other plot parameters
plt.figure(figsize = (15,10))
plt.title('Tumor Volume Change over 45 Day Treatment', fontdict = {'fontsize': 20, 'fontweight': 'bold'})
plt.ylabel('Survival Rate (%)')
plt.axhline(y=0, color = 'black') #adds a horizontal line at zero


plt.bar(xaxis, tumor_change, tick_label=drug,
       color = ['red' if tumor_change[y] > 0 else 'green' for y in range (0,x)])

# Ploting text string
for i in range (0, x):
    if tumor_change[i] > 0:
        plt.text(i+1,5, str(f"{tumor_change[i]}%") , fontsize=12, color='black', ha = 'center')
    else:
        plt.text(i+1,-5, str(f"{tumor_change[i]}%") , fontsize=12, color='white', ha = 'center')      
        

# Setting theme using Seaborne library function
sns.set_style('ticks')

plt.savefig("Images/Chart-4-BarGraph_Tumor_Volume_Change.png")

plt.show()
```
